```python
import pandas as pd
import re
import numpy as np
import re
```

```python
finbert_tweet = pd.read_csv('finbert_tweet.csv')
finbert_tweet
```

# Sum with likes

```python
finbert_likes_sum = finbert_tweet.copy()
finbert_likes_sum['positive'] = finbert_likes_sum.Likes * finbert_likes_sum.positive
finbert_likes_sum['negative'] = finbert_likes_sum.Likes * finbert_likes_sum.negative
finbert_likes_sum['neutral'] = finbert_likes_sum.Likes * finbert_likes_sum.neutral
finbert_likes_sum
```

```python
finbert_likes_sum = finbert_likes_sum[['date', 'positive', 'negative', 'neutral']]
finbert_likes_sum = finbert_likes_sum.groupby('date').sum()
finbert_likes_sum
```

# Average without likes

```python
finbert_avg = finbert_tweet.copy()
finbert_avg = finbert_avg[['date', 'positive', 'negative', 'neutral']]
finbert_avg = finbert_avg.groupby('date').mean()
finbert_avg
```

```python
finbert_likes_sum.to_csv('finbert_likes_sum.csv')
finbert_avg.to_csv('finbert_avg.csv')
```
