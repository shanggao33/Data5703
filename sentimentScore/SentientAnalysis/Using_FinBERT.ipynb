```python
!pip install yfinance
import numpy as np 
import pandas as pd 
from tqdm import tqdm
import datetime
import matplotlib.pyplot as plt
import tensorflow as tf
import torch
import re
import yfinance as yf
import holidays
import pickle
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers.experimental import preprocessing
!pip install transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
import torch.nn.functional as F
```

```python
data_NWSA.dropna(inplace = True)
data_NWSA.reset_index(drop=True, inplace=True)
data_NWSA.rename(columns={'release_date':'date'}, inplace=True)
data_NWSA.drop(inplace=True, columns=['category', 'provider', 'url', 'article_id'], axis=1)
```

```python
data_NWSA.head()
```

```python
def SentimentAnalyzer(doc):
    pt_batch = tokenizer(doc,padding=True,truncation=True,max_length=512,return_tensors="pt")
    outputs = model(**pt_batch)
    pt_predictions = F.softmax(outputs.logits, dim=-1)
    return pt_predictions.detach().cpu().numpy()
```

```python
tmpData = {}
total = data_NWSA['date'].nunique()
for i in tqdm(data_NWSA['date'].unique()):
    tmpData[i] = data_NWSA.loc[data_NWSA['date'] == i]['title'].tolist()
```

```python
tmpData['2016-11-06']
```

```python
def findPercentageBySentences(sentenceList):
    posAvg, negAvg, neuAvg = 0, 0, 0
    sentimentArr = SentimentAnalyzer(sentenceList)
    sentimentArr = np.mean(sentimentArr, axis=0)
    posAvg=sentimentArr[0]
    negAvg=sentimentArr[1]
    neuAvg=sentimentArr[2]
    return {'numArticles': len(sentenceList), 'pos': posAvg, 'neg': negAvg, 'neu' : neuAvg}
```

```python
dateSentimentGroups = {}
for i in tqdm(tmpData):
    scores = findPercentageBySentences(tmpData[i])
    dateSentimentGroups[i] = scores
```

```python
dateSentimentGroups['2016-11-06']
```

```python
ONE_DAY = datetime.timedelta(days=1)
HOLIDAYS_US = holidays.US()
def next_business_day(dateString):
    datetimeObj = datetime.datetime.strptime(dateString, '%Y-%m-%d')
    next_day = datetimeObj + ONE_DAY
    while next_day.weekday() in holidays.WEEKEND or next_day in HOLIDAYS_US:
        next_day += ONE_DAY
    return next_day
```

```python
label = []
pred = []
NWSA = []
ticker = yf.Ticker('NWSA')
hist = ticker.history(period="max")
for i in tqdm(dateSentimentGroups):
  start = i
  nextDay = next_business_day(start).strftime("%Y-%m-%d")
  NWSA.append([i,dateSentimentGroups[i]['numArticles'], dateSentimentGroups[i]['neu'], dateSentimentGroups[i]['pos'], dateSentimentGroups[i]['neg']])
  try:
    curDay = hist.loc[start]
    #nextDay = hist.loc[nextDay]
    if curDay['Close'] >= curDay['Open']:
      label.append(1)
    elif curDay['Close'] < curDay['Open']:
      label.append(-1)
    if dateSentimentGroups[i]['pos'] >= dateSentimentGroups[i]['neg']:
      pred.append(1)
    elif dateSentimentGroups[i]['pos'] < dateSentimentGroups[i]['neg']:
      pred.append(-1)
  except:
    pass
# the data can not load matched label
```

```python
df1 = pd.DataFrame(columns =['data','numArticles', 'neutral', 'positive','negative'], data=NWSA)
df1.to_csv('/content/gdrive/MyDrive/COMP5703/week7/SentientAnalysis/raw data/bert_nwsa.csv')
```

```python
df1
```

```python
from sklearn.metrics import classification_report
print(classification_report(label, pred))
```
