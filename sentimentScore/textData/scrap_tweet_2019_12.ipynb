```python
from Scweet.scweet import scrape

def scrap_by_month(since, until, filename):
    data = scrape(hashtag="BAC", since=since, until=until, from_account=None, interval=1, lang='en',
                headless=True, display_type="Latest", save_images=False,  save_dir=filename,
                resume=False, proximity=False)
    
scrap_by_month("2011-1-1", "2012-1-1", "2011") ## scrap by year actually. Change the year to cover 2010 to 2016
```
