```python
import os
file_dir = '/home/sean/Data5703/dataset/bac/8combinations'
file_list = os.listdir(file_dir)

for file in file_list:
    df_indi = pd.read_csv('/home/sean/Data5703/dataset/bac/bac_full_indicators.csv')
    df_sent = pd.read_csv(file_dir + '/' + file)

    # drop open, high, low, close, volume in df_sent
    df_sent = df_sent.drop(['open', 'high', 'low', 'close', 'volume'], axis=1)

    # merge df_sent and df_indi
    df = pd.merge(df_sent, df_indi, on='date', how='inner')

    # save to csv
    df.to_csv(file_dir + '/indi_' + file, index=False)
    
```
